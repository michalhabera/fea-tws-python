Plot contour for function
$$f(x, y) = (1 - \frac{x}{2} + x^5 + y^3) \exp(-x^2 - y^2)$$
between $(-3, 3) \times (-3, 3)$.

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y)

def f(x, y):
    return (1.0 - x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)


plt.figure(figsize=(6, 3))
cbar = plt.contourf(X, Y, f(X, Y), 20, cmap="viridis")
C = plt.contour(X, Y, f(X, Y), 10, colors="black")
plt.clabel(C, inline=True, fontsize=8)
plt.colorbar(cbar)
```
